```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau = 500.

x = np.arange(0,9000)

y1 = np.ones(3000)*9
y2 = np.ones(3000)*8
y3 = 0.5*np.exp(-np.arange(0,3000)/tau)+7.5


y = np.concatenate([y1,y2,y3])

plt.plot(x,y)

# First shift 
plt.arrow(2800, 8, 0, 1, color='red', head_length = 0.1, head_width = 100, length_includes_head = True)
plt.arrow(2800, 9, 0, -1, color='red', head_length = 0.1, head_width = 100, length_includes_head = True)
plt.text(2100, 8.5, '$\Delta f_L$', fontsize = 13)

# Second shift 
plt.arrow(5800, 7.5, 0, 0.5, color='red', head_length = 0.1, head_width = 100, length_includes_head = True)
plt.arrow(5800, 8, 0, -0.5, color='red', head_length = 0.1, head_width = 100, length_includes_head = True)
plt.text(5100, 7.75, '$\Delta f_m$', fontsize = 13)


plt.yticks([9, 8, 7.5],['$f_0$', '$f_L$', '$f_m$'])
plt.xticks([0, 3000, 6000], ['$t_0$', '$t_1$', '$t_2$'])

plt.title('Ideal chase')
plt.show()
```
